```python
import tkinter as tk
from tkinter import filedialog
import pandas as pd
import datetime
import time
```

```python
window = tk.Tk()
window.title('Karn Naphob Data Extraction')
```

```python
def open_file():
    file_path = filedialog.askopenfilename()
    print("Choose File:", file_path)

    # Load the data
    data_load = pd.read_excel(file_path)
    print(data_load)

    # Process the data
    station = []
    time_save = []
    pr_save = []
    sta = -1
    for i in range(len(data_load)):
        if type(data_load.loc[i][1]) == str:
            if len(data_load.loc[i][1]) > 5:
                check_station = station.count(data_load.loc[i][1])
                if check_station == 0:
                    station.append(data_load.loc[i][1])
                    time_save.append([])
                    pr_save.append([])
                    sta += 1
                m = data_load.loc[i][2].month
                day_now = data_load.loc[i][2]
                d = m + 1
                del_time = datetime.timedelta(days=1)
                if d > 12:
                    d -= 12
                num = 0
                while m != d:
                    time_save[sta].append(day_now)
                    pr_save[sta].append(data_load.loc[i][3 + num])
                    day_now += del_time
                    m = day_now.month
                    num += 1

    # Save the extracted data
    save = pd.DataFrame()
    for i in range(len(station)):
        save[station[i] + ' time'] = pd.Series(time_save[i])
        save[station[i]] = pd.Series(pr_save[i])
    save.to_excel('rainy.xlsx')

    label.config(text="File Selected: " + file_path)

Label_karn = tk.Label(text='By Karn')
Label_karn.pack()
label = tk.Label(window, text="")
label.pack()

button = tk.Button(window, text="Please Select File", command=open_file)
button.pack()

# Show window
window.mainloop()
```
