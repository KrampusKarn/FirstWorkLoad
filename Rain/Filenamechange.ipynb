```python
import pandas as pd
import numpy as np

import os
import warnings
import re
```

```python

excel_path = 'C:/Users/Windows10/Desktop/Documents/ร่าง 68(ทิ้ง)(ล่าสุด).xlsx'
df = pd.read_excel(excel_path)

pass_dict = pd.Series(df.ID.values, index=df.รหัส.astype(str)).to_dict()
print("pass_dict:", pass_dict)

directory_path = 'D:/ร่างกานต์ 68/สปช.1/Raw16'
files = os.listdir(directory_path)
```

```python
# Counter for renamed files
renamed_count = 0
# Step 3: Rename files
for filename in files:
    file_path = os.path.join(directory_path, filename)
    
    # Check if it's a file (not a directory)
    if os.path.isfile(file_path):
        renamed = False  # Flag to check if file is renamed
        
        # Debugging: Print current filename
        print(f'Processing file: {filename}')
        
        # Extract any 'รหัส' part from the filename using regex
        # The regex looks for patterns like '68-05-23' or '68-06-1005' which is common for รหัส
        match = re.search(r'\b(\d{2}-\d{2}-\d{2,4})\b', filename)
        
        if match:
            รหัส = match.group(1)
            # Debugging: Print extracted รหัส
            print(f'Extracted รหัส: {รหัส}')
            
            # Check if the รหัส exists in pass_dict
            if รหัส in pass_dict:
                id_value = pass_dict[รหัส]
                new_filename = f"{id_value}{os.path.splitext(filename)[1]}"
                new_file_path = os.path.join(directory_path, new_filename)
                
                # Check if the new filename already exists and skip if it does
                if os.path.exists(new_file_path):
                    print(f'Skipped "{filename}" because "{new_filename}" already exists')
                    renamed = True
                else:
                    # Rename the file
                    os.rename(file_path, new_file_path)
                    print(f'Renamed "{filename}" to "{new_filename}"')
                    renamed = True
                    renamed_count += 1  # Increment the counter

                    # Update the Excel DataFrame to mark 'มี'
                    df.loc[df.รหัส == รหัส, 'มี/ไม่มี'] = 'มี'
                    print(f'Updated Excel for รหัส: {รหัส}')
            else:
                print(f'Skipped "{filename}" (no matching ID found for รหัส "{รหัส}")')
        else:
            print(f'Skipped "{filename}" (no valid รหัส found)')

# Display the total number of renamed files
print(f'Total files renamed: {renamed_count}')
```

```python
# Load the Excel file and create the pass_dict dictionary
excel_path = 'C:/Users/Windows10/Desktop/Documents/ร่าง 68(ทิ้ง)(ล่าสุด).xlsx'
df = pd.read_excel(excel_path)

# Creating the dictionary from the Excel file
pass_dict = pd.Series(df.ID.values, index=df.รหัส.astype(str)).to_dict()
print("pass_dict:", pass_dict)  # Debug: Print pass_dict

# Directory to traverse (assuming this is the base directory to start from)
base_directory = 'Z:/งบประมาณปี2568/pdf ปมก. 68/สชป.13'
```

```python
# Initialize counter for renamed files and list to keep track of renamed IDs
renamed_count = 0
renamed_ids = []

# Function to match รหัส patterns in folder names
def extract_รหัส(foldername):
    match = re.match(r'68-\d{1,2}-\d{1,4}', foldername)
    if match:
        return match.group(0)
    return None

# Traverse the base directory
for root, dirs, files in os.walk(base_directory):
    for dir_name in dirs:
        รหัส = extract_รหัส(dir_name)
        if รหัส:
            full_dir_path = os.path.join(root, dir_name)
            print(f"Found directory: {full_dir_path} with รหัส: {รหัส}")  # Debug: Print directory and รหัส
            if รหัส in pass_dict:
                id_value = pass_dict[รหัส]
                file_path = os.path.join(full_dir_path, '0.pdf')
                if os.path.exists(file_path):
                    new_file_path = os.path.join(full_dir_path, f"{id_value}.pdf")
                    if os.path.exists(new_file_path):
                        print(f'Skipped "{file_path}" because "{new_file_path}" already exists')
                    else:
                        os.rename(file_path, new_file_path)
                        print(f'Renamed "{file_path}" to "{new_file_path}"')
                        renamed_count += 1  # Increment the counter
                        renamed_ids.append(id_value)  # Track the renamed ID
                else:
                    print(f'File "0.pdf" does not exist in {full_dir_path}')
            else:
                print(f'รหัส "{รหัส}" not found in pass_dict')

# Display the total number of renamed files
print(f'Total files renamed: {renamed_count}')

# Group the renamed IDs and display the summary
unique_ids = set(renamed_ids)
print("Summary of renamed IDs:")
for id_value in unique_ids:
    count = renamed_ids.count(id_value)
    print(f'ID {id_value} was used to rename {count} files')
```

```python
import shutil
base_directory = r'Z:\งบประมาณปี2568\pdf ปมก. 68\สชป.13'
# Target directory to move files to
target_directory = r'D:\ร่างกานต์ 68\สปช.1\ปมกpdfraw'

# Create the target directory if it doesn't exist
os.makedirs(target_directory, exist_ok=True)

# List to keep track of files with two or more digit names
renamed_files = []

# Function to match filenames with two or more digits
def is_two_or_more_digits(filename):
    match = re.match(r'^\d{2,}\.pdf$', filename)
    return match is not None

# Traverse the base directory
for root, dirs, files in os.walk(base_directory):
    for file_name in files:
        if is_two_or_more_digits(file_name):
            full_file_path = os.path.join(root, file_name)
            renamed_files.append(full_file_path)

# Move the files to the target directory
for file_path in renamed_files:
    # Extract the file name from the full path
    file_name = os.path.basename(file_path)
    # Define the target file path
    target_file_path = os.path.join(target_directory, file_name)
    # Move the file
    shutil.move(file_path, target_file_path)
    print(f'Moved "{file_path}" to "{target_file_path}"')

# Display the total number of moved files
print(f'Total files moved: {len(renamed_files)}')
```
