```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
East_af = pd.read_csv("EasternAFRI.csv")
```

```python
East_af
```

```python
East_af_new = East_af.drop(index=range(24, 29)).drop(columns="Series Code")
```

```python
East_af_new
```

```python
East_africa = pd.melt(East_af_new, id_vars=['Country Name', 'Country Code', 'Series Name'], var_name='Year', value_name='Value')

# Extract the year from the 'Year' column
East_africa['Year'] = East_africa['Year'].str.extract(r'(\d+)').astype(int)
```

```python
East_africa
```

```python
East_africa = East_africa.pivot(index=['Country Name', 'Country Code', 'Year'], columns='Series Name')['Value'].reset_index()
East_africa.columns.name = None
East_africa
```

```python
East_africa.head(
)
```

```python
East_africa = East_africa.sort_values(by=['Year', 'Country Name'])

# Reset index (optional)
East_africa.reset_index(drop=True)
```

```python
East_africa['Year'] = East_africa['Year'].astype(str)

# Convert 'GDP growth (annual %)' and 'Trade (% of GDP)' columns to float
East_africa['GDP growth (annual %)'] = East_africa['GDP growth (annual %)'].astype(float)
East_africa['Trade (% of GDP)'] = East_africa['Trade (% of GDP)'].astype(float)
```
